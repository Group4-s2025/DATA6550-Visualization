```python
def pass_id(id):
    if id < 100:
        return '#1-99'
    elif 100 <= id < 200:
        return "#100-199 "
    elif 200 <= id < 300:
        return "#200-299"
    elif 300 <= id < 400:
        return "#300-399"
    elif 400 <= id < 500:
        return "#400-499"
    elif 500 <= id < 600:
        return "#500-599 "
    elif 600 <= id < 700:
        return "#600-699"
    elif 700 <= id < 800:
        return "#700-799"
    elif 800 <= id < 900:
        return "#800-899 "
    else:
        return "over 900"

train['id'] = train['PassengerId'].apply(pass_id)
print(train['id'].value_counts())

frequencies = train['id'].value_counts()



plt.bar(frequencies.index, frequencies.values)


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('')


plt.show()
```

```python

plt.figure(figsize=(12, 5))
plt.bar(frequencies.index, frequencies.values)
plt.yticks(np.arange(100, 1000, step=100))

plt.xlabel('Passenger ID categories')
plt.ylabel('Count')
plt.title('Number of passenger IDs by category')


plt.show()
```

```python
def pass_id(id):
    if id < 100:
        return '#1-99'
    elif 100 <= id < 200:
        return "#100-199 "
    elif 200 <= id < 300:
        return "#200-299"
    elif 300 <= id < 400:
        return "#300-399"
    elif 400 <= id < 500:
        return "#400-499"
    elif 500 <= id < 600:
        return "#500-599 "
    elif 600 <= id < 700:
        return "#600-699"
    elif 700 <= id < 800:
        return "#700-799"
    elif 800 <= id < 900:
        return "#800-899 "
    else:
        return "over 900"

train['id'] = train['PassengerId'].apply(pass_id)
print(train['id'].value_counts().sort_index(ascending=True))

frequencies = train['id'].value_counts().sort_index()



plt.bar(frequencies.index, frequencies.values)
plt.yticks(np.arange(100, 1000, step=100))

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('')


plt.show()
plt.figure(figsize=(12, 5))
plt.bar(frequencies.index, frequencies.values)

#custom_order = train['id'].to_list()
#custom_order_unique = list(set(custom_order))
#plt.xticks(ticks=range(len(custom_order_unique)), labels=custom_order_unique )
plt.xlabel('Passenger ID categories')
plt.ylabel('Count')
plt.title('Number of passenger IDs by category')


plt.show()
```
