```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
url1 = "https://raw.githubusercontent.com/Group4-s2025/DATA6550-Visualization/refs/heads/main/Data/gender_submission.csv"
url2 = "https://raw.githubusercontent.com/Group4-s2025/DATA6550-Visualization/refs/heads/main/Data/test.csv"
url3 = "https://raw.githubusercontent.com/Group4-s2025/DATA6550-Visualization/refs/heads/main/Data/train.csv"
gender = pd.read_csv(url1)
test = pd.read_csv(url2)
train = pd.read_csv(url3)
```

# Good Visualizations

```python
# Survival by class
sns.countplot(x='Pclass', hue='Survived', data=train)
plt.title('Survival Rate by Passenger Class'), plt.xlabel('Passenger Class'), plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2], labels=['First', 'Second', 'Third'])
plt.show()
```

```python
# survival ratios by class
survival_ratio = train.groupby('Pclass')['Survived'].mean().reset_index()

# graph the survival ratios
sns.barplot(x='Pclass', y='Survived', data=survival_ratio, palette='viridis')
plt.title('Survival Ratio by Passenger Class'), plt.xlabel('Passenger Class'), plt.ylabel('Survival Ratio')
plt.xticks(ticks=[0, 1, 2], labels=['First', 'Second', 'Third'])
plt.ylim(0, 1)
plt.show()
```

```python
# Survival by gender
sns.countplot(x='Sex', hue='Survived', data=train)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# survival ratios by gender
survival_ratio_gender = train.groupby('Sex')['Survived'].mean().reset_index()

# graph the survival ratios
sns.barplot(x='Sex', y='Survived', data=survival_ratio_gender, palette='viridis')
plt.title('Survival Ratio by Gender'), plt.xlabel('Gender'), plt.ylabel('Survival Ratio')
plt.ylim(0, 1)
plt.show()
```

```python
# Create age groups
train['age_group'] = pd.cut(train['Age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior'])

# Survival by age group
sns.countplot(x='age_group', hue='Survived', data=train)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# survival ratios by age group
survival_ratio_age = train.groupby('age_group')['Survived'].mean().reset_index()

# graph the survival ratios
sns.barplot(x='age_group', y='Survived', data=survival_ratio_age, palette='viridis')
plt.title('Survival Ratio by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Ratio')
plt.ylim(0, 1)
plt.show()
```

```python
# Survival by embarkation point
sns.countplot(x='Embarked', hue='Survived', data=train)
plt.title('Survival Rate by Embarkation Point')
plt.xlabel('Embarked Location')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate survival ratios by embarkation point
survival_ratio_embarked = train.groupby('Embarked')['Survived'].mean().reset_index()

# Plot the survival ratios
sns.barplot(x='Embarked', y='Survived', data=survival_ratio_embarked, palette='viridis')
plt.title('Survival Ratio by Embarkation Point')
plt.xlabel('Embarked Location')
plt.ylabel('Survival Ratio')
plt.ylim(0, 1)
plt.xticks(ticks=[0, 1, 2], labels=['Cherbourg', 'Queenstown', 'Southampton'])
plt.show()

```

```python
# average fare by embarkation point
avg_fare_embarked = train.groupby('Embarked')['Fare'].mean().reset_index()

# graph the average fares
sns.barplot(x='Embarked', y='Fare', data=avg_fare_embarked, palette='magma')
plt.title('Average Ticket Fare by Embarkation Point')
plt.xlabel('Embarked Location')
plt.ylabel('Average Fare')
plt.xticks(ticks=[0, 1, 2], labels=['Cherbourg', 'Queenstown', 'Southampton'])
plt.show()
```

```python
# Create fare groups
train['fare_group'] = pd.cut(train['Fare'], bins=[0, 50, 100, 150, 200, 1000], labels=['<50', '50-100', '100-150', '150-200', '>200'])

# Survival by fare group
sns.countplot(x='fare_group', hue='Survived', data=train)
plt.title('Survival Rate by Fare')
plt.xlabel('Fare Range')
plt.ylabel('Count')
plt.show()
```
